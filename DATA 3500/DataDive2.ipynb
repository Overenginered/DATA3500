```python
# 7. Degrees with the Largest Salary Growth from Starting to Mid-Career
# Calculate the salary growth from starting to mid-career for each degree
degrees_pay_back['Salary Growth (%)'] = ((degrees_pay_back['Mid-Career Median Salary'] - degrees_pay_back['Starting Median Salary']) / degrees_pay_back['Starting Median Salary']) * 100

# Find the degrees with the largest salary growth
largest_salary_growth_degrees = degrees_pay_back[['Undergraduate Major', 'Salary Growth (%)']].sort_values(by='Salary Growth (%)', ascending=False).head(10)

# Plot the degrees with the largest salary growth
plt.figure(figsize=(10, 6))
sns.barplot(x='Salary Growth (%)', y='Undergraduate Major', data=largest_salary_growth_degrees)
plt.title('Degrees with the Largest Salary Growth from Starting to Mid-Career')
plt.xlabel('Salary Growth (%)')
plt.ylabel('Undergraduate Major')
plt.show()
```

```python
#8. Cost of Living Adjustment for Different Regions
# Hypothetical cost of living indices
cost_of_living_indices = {
    'California': 150,
    'Northeastern': 130,
    'Southern': 90,
    'Western': 110,
    'Midwestern': 85
}

# Adjust the starting median salaries by the cost of living index
salaries_by_region['Cost of Living Index'] = salaries_by_region['Region'].map(cost_of_living_indices)
salaries_by_region['Adjusted Starting Median Salary'] = salaries_by_region['Starting Median Salary'] / (salaries_by_region['Cost of Living Index'] / 100)

# Calculate the average adjusted starting median salary by region
average_adjusted_salary_by_region = salaries_by_region.groupby('Region')['Adjusted Starting Median Salary'].mean().sort_values(ascending=False).reset_index()

# Plot the adjusted starting median salaries by region
plt.figure(figsize=(10, 6))
sns.barplot(x='Adjusted Starting Median Salary', y='Region', data=average_adjusted_salary_by_region)
plt.title('Average Adjusted Starting Median Salary by Region')
plt.xlabel('Adjusted Starting Median Salary ($)')
plt.ylabel('Region')
plt.show()
```
