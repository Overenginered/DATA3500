```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the provided CSV files
corruption_df = pd.read_csv('corruption.csv')
cost_of_living_df = pd.read_csv('cost_of_living.csv')
richest_countries_df = pd.read_csv('richest_countries.csv')
tourism_df = pd.read_csv('tourism.csv')
unemployment_df = pd.read_csv('unemployment.csv')
demographics_df = pd.read_csv('final_demographics_data.csv')
worldometer_df = pd.read_csv('worldometer_data.csv')

# Set seaborn style
sns.set(style="whitegrid")

# 1. Corruption and COVID-19 case fatality rate
fig1, ax1 = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=worldometer_df.merge(corruption_df, left_on='Country/Region', right_on='country'),
                x='corruption_index', y='TotalDeaths', hue='corruption_index', palette='viridis', ax=ax1)
ax1.set_title('Corruption Index vs. Total COVID-19 Deaths')
ax1.set_xlabel('Corruption Index')
ax1.set_ylabel('Total COVID-19 Deaths')
```

```python
# 2. Cost of living and economic impact
fig2, ax2 = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=cost_of_living_df.merge(worldometer_df, left_on='country', right_on='Country/Region'),
                x='cost_index', y='TotalCases', hue='cost_index', palette='coolwarm', ax=ax2)
ax2.set_title('Cost of Living Index vs. Total COVID-19 Cases')
ax2.set_xlabel('Cost of Living Index')
ax2.set_ylabel('Total COVID-19 Cases')
```

```python
# 3. GDP per capita and pandemic response effectiveness
fig3, ax3 = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=richest_countries_df.merge(worldometer_df, left_on='country', right_on='Country/Region'),
                x='gdp_per_capita', y='TotalCases', hue='gdp_per_capita', palette='plasma', ax=ax3)
ax3.set_title('GDP per Capita vs. Total COVID-19 Cases')
ax3.set_xlabel('GDP per Capita')
ax3.set_ylabel('Total COVID-19 Cases')
```

```python
# 4. Impact of decline in tourism
fig4, ax4 = plt.subplots(figsize=(10, 6))
sns.barplot(data=tourism_df.merge(worldometer_df, left_on='country', right_on='Country/Region'),
            x='country', y='tourists_in_millions', palette='Set3', ax=ax4)
ax4.set_title('Impact of Decline in Tourism')
ax4.set_xlabel('Country')
ax4.set_ylabel('Tourists in Millions')
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=90)
```

```python
# 5. Unemployment rate and government interventions
fig5, ax5 = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=unemployment_df.merge(worldometer_df, left_on='country', right_on='Country/Region'),
                x='unemployment_rate', y='TotalCases', hue='unemployment_rate', palette='icefire', ax=ax5)
ax5.set_title('Unemployment Rate vs. Total COVID-19 Cases')
ax5.set_xlabel('Unemployment Rate')
ax5.set_ylabel('Total COVID-19 Cases')
```

```python
# 6. Demographic factors and COVID-19 spread
fig6, ax6 = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=demographics_df.merge(worldometer_df, left_on='country', right_on='Country/Region'),
                x='Urban population', y='TotalCases', hue='Urban population', palette='viridis', ax=ax6)
ax6.set_title('Urban Population vs. Total COVID-19 Cases')
ax6.set_xlabel('Urban Population')
ax6.set_ylabel('Total COVID-19 Cases')
```

```python
# 7. Corruption index and COVID-19 mortality
fig7, ax7 = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=worldometer_df.merge(corruption_df, left_on='Country/Region', right_on='country'),
                x='corruption_index', y='Deaths/1M pop', hue='corruption_index', palette='coolwarm', ax=ax7)
ax7.set_title('Corruption Index vs. COVID-19 Deaths per 1M Population')
ax7.set_xlabel('Corruption Index')
ax7.set_ylabel('COVID-19 Deaths per 1M Population')
```

```python
# 8. Cost of living and food insecurity
fig8, ax8 = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=cost_of_living_df.merge(worldometer_df, left_on='country', right_on='Country/Region'),
                x='cost_index', y='TotalDeaths', hue='cost_index', palette='plasma', ax=ax8)
ax8.set_title('Cost of Living Index vs. Total COVID-19 Deaths')
ax8.set_xlabel('Cost of Living Index')
ax8.set_ylabel('Total COVID-19 Deaths')
```

```python
# 9. Wealth and pandemic response speed
fig9, ax9 = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=richest_countries_df.merge(worldometer_df, left_on='country', right_on='Country/Region'),
                x='gdp_per_capita', y='TotalRecovered', hue='gdp_per_capita', palette='viridis', ax=ax9)
ax9.set_title('GDP per Capita vs. Total COVID-19 Recoveries')
ax9.set_xlabel('GDP per Capita')
ax9.set_ylabel('Total COVID-19 Recoveries')
```

```python
# 10. Testing rates and COVID-19 cases
fig10, ax10 = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=worldometer_df,
                x='Tests/1M pop', y='TotalCases', hue='Tests/1M pop', palette='inferno', ax=ax10)
ax10.set_title('Testing Rates vs. Total COVID-19 Cases')
ax10.set_xlabel('Tests per 1M Population')
ax10.set_ylabel('Total COVID-19 Cases')

plt.tight_layout()
plt.show()
```
